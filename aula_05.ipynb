### Passo 1 -> importar as bilbiotecas

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
```

plotando um mapa simples com o basemap (apenas o contorno dos continentes)

```python
m = Basemap() # define um objeto do tipo basemap
m.drawcoastlines()# plota uma linha de costa
plt.show()# mostra a imagem
```

### plotando um mapa colorido da região antartica

```python
m = Basemap(projection='splaea',boundinglat=-10,lon_0=90,resolution='l')
m.drawcoastlines()
m.fillcontinents(color='coral',lake_color='aqua')
# plotando os paralelos e meridianos
m.drawparallels(np.arange(-80.,81.,20.))
m.drawmeridians(np.arange(-180.,181.,20.))
m.drawmapboundary(fill_color='aqua') # preenchendo os contornos do mapa com a cor "aqua"
```

### Projecao Cilindrica

Cylindrical Equidistant projection

o lon_0 e o lat_0 refer to the center of the map!

```python
m = Basemap(projection= 'aeqd',lon_0=-74, lat_0=40);# defining a map with equidistant projection
m.drawmapboundary(fill_color='lightblue');#outline of the oceans with the color blue
m.fillcontinents(color= 'Darkorange', lake_color='skyblue');# coloring the continents orange
plt.show() 
```

### Mapa comprojeção Cilindrica

```python
# setando o tamanho da figura
plt.figure(figsize=(20,8))

m = Basemap(projection='cyl',lon_0=0,
            lat_0=0,lat_ts=0,
            llcrnrlat=-90,urcrnrlat=90,
            llcrnrlon=-180,urcrnrlon=180)
m.drawcoastlines();
m.drawmapboundary(fill_color='lightblue')
m.fillcontinents(color='darkorange')# by coloring the continents
m.drawcountries()

m.drawmeridians(range(-180,180,30),
                labels=[0,0,1,]);
m.drawparallels(range(-80,90,15),
                labels=[1,0,0])
plt.show()
```

### plotando pontos em um papa

```python
lon, lat = -38, -3 # declarando as coordenadas
x, y = m(lon, lat) # transformando as coordenadas em pixels
```

```python
plt.figure(figsize=(20,8))
m = Basemap(projection='cyl',lon_0=0,lat_0=0,lat_ts=-3,
            llcrnrlat=-60,urcrnrlat=35,
            llcrnrlon=-85,urcrnrlon=30,
            resolution='c')
m.drawcoastlines();
m.drawmapboundary(fill_color='skyblue')
m.fillcontinents(color='darkorange')
m.drawcountries()
m.plot(x,y,marker='s',markersize=12,color='black')
m.drawmeridians(np.arange(-180,180,15),labels=[0,0,1,]);
m.drawparallels(np.arange(-90,90,15),labels=[1,0,0])
plt.show()
```
